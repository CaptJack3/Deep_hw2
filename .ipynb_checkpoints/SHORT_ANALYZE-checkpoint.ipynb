```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%load_ext autoreload
%autoreload 2

plt.rcParams.update({'font.size': 12})
# test = unittest.TestCase()
print("Finish")
```

```python
''' READ DATA FILE '''
data_file_name='Jan2Oct_Short.xlsx'
data=pd.read_excel(data_file_name)
print("Finish")
```

```python
# Print the results per month and Direction (Short,Long)"
df2 = data.groupby(['Month','Dir'])['P/L'].sum()
print("Get sum of the grouped data By Month and Dir:\n", df2)
df2 = data.groupby(['Month'])['P/L'].sum()
print("Get sum of the grouped data By Month Only:\n", df2)
print(f"\n Total result of the analysed data is: {data['P/L'].sum()}\n")
print("Finish")
```

```python
# Plot the Nasdaq Mid Value over the tested time:
plt.plot(data['Date'],data['Mid'])
plt.title('Nasdaq Mid Value Over time')
plt.xlabel('date')
plt.ylabel('Mid')
plt.show()
```

```python
# For Time Analysis 
'''TIME ANALYSIS'''
data['date'] = pd.to_datetime(data['Date'])
# change the datetime format
data['time'] = data['date'].dt.strftime('%H:%M')
data['time_H'] =  data['date'].dt.strftime('%H').astype(int)
data['time_M'] =  data['date'].dt.strftime('%M').astype(int)
# Transform day time to minute scale. 16:00 defined as the begining of the day, i.e. t=0
data['t'] = 60 * ( (data['time_H']) - 16 ) + (data['time_M'])

'''Checking The results'''
# print(data['time'])
# print("********************")
# print(data['date'])
# print("********************")
# print(data['time_H'])
# print("********************")
# print(data['time_M'])
# print("********************")
# print(data['t'])
# print("********************")
print("Finish")
```

```python
# SCATTER Visualization of trade's P/L as function of t (minutes starting from 16:00)
sns.scatterplot(y='t', x='P/L', data = data)
# data.plot.scatter(x="Vol", y="P/L")
plt.title('P/L vs Time ')
plt.ylabel('minutes from 16:00')
plt.xlabel('P/L')
plt.grid()
plt.show()
print("Finish")
```

```python
# SCATTER Visualization of trade's P/L as function of t (minutes starting from 16:00)
# Bivariate analysis:
sns.histplot(data, x="t", y="P/L")
plt.title('Short trades density vs P/L and time[min] from 16:00')
# Best Times to trade:
# plt.axvline(60, 0,1,c='r')
# plt.axvline(120, 0,1,c='r')
# plt.axvline(210, 0,1,c='r')
# plt.axvline(270, 0,1,c='r')
plt.show()
```

```python
# Does there is a significant change in the vol as function of month?
# Plot the Nasdaq Mid Value over the tested time:

# plt.plot(data['Date'],data['Mid'])
# plt.xlabel('date')
# # Plot the Vol:
# plt.plot(data['Date'],data['Vol'],'.')
# plt.show()

##########
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Mid', color=color)
ax1.plot(data['Date'],data['Mid'], color=color)
ax1.tick_params(axis='y', labelcolor=color)
#
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('Vol', color=color)  # we already handled the x-label with ax1
ax2.plot(data['Date'],data['Vol'],'.', color=color)
ax2.tick_params(axis='y', labelcolor=color)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

```python
# Cumulative Sum of PL
P_L_cs =np.cumsum(data['P/L'])
plt.plot(P_L_cs)
plt.show()
```

```python
# Time Analysis
time_analysis_df = data[['P/L','Vol','t']]
time_analysis_df =time_analysis_df.sort_values(by='t')
# time_analysis_df['t']
# plt.plot(time_analysis_df['t'],time_analysis_df['P/L'],'.')
# plt.plot(time_analysis_df['t'],np.cumsum(time_analysis_df['P/L']))
# plt.xlabel('min from 16:00')
###########

##########
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel('min from 16:00')
ax1.set_ylabel('cum. sum', color=color)
ax1.plot(time_analysis_df['t'],np.cumsum(time_analysis_df['P/L']), color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.axvline(60,0,1)
ax1.axvline(150,0,1)
ax1.axvline(270,0,1)
ax1.axvline(360,0,1)
#
ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
color = 'tab:blue'
ax2.set_ylabel('P/L', color=color)  # we already handled the x-label with ax1
ax2.plot(time_analysis_df['t'],time_analysis_df['P/L'],'.', color=color,ms=0.8)
ax2.tick_params(axis='y', labelcolor=color)
fig.tight_layout()  # otherwise the right y-label is slightly clipped
# plt.show()
ax1.grid()
```

```python
# trade_length_min = 
# data.assign('trade_length_min')
trade_length_list = list()
for ii in range(0,len(data)):
    temp = 60 * data['position time'][ii].hour + data['position time'][ii].minute + (data['position time'][ii].second/60)
    
    # print(data['position time'][ii],temp)
    # data['trade_length_min'][ii] = temp
    trade_length_list.append(temp)
data['trade_length_list'] = np.round(np.array(trade_length_list),1)



# print("finish")
data['trade_length_list']

# plt.plot(data['trade_length_list'],data['P/L'],'.')
# plt.xlabel('trade length [min]')
# plt.ylabel('PL')

df_time_length = data[['trade_length_list','P/L']]
df_time_length = df_time_length.sort_values(by='trade_length_list')
plt.plot(df_time_length['trade_length_list'],np.cumsum(df_time_length['P/L']))
plt.xlabel('trade length [min]')
plt.ylabel('cumulative PL')

```
